```python
import numpy as np
import pandas as pd
```

```python
def get_most_common(year):
    df = pd.read_csv(f"songs_cleaned/songs_{year}.csv", sep=";")
    words = []
    for i, lyrics in enumerate(df["lyrics"]):
        for word in lyrics.split(" "):
            # print(word)
            words.append(word)
    words, counts = np.unique(words, return_counts=True)
    words = [x for _, x in sorted(zip(counts, words))][::-1]
    counts = sorted(counts)[::-1]
    # for i in range(len(words)):
    counter = 0
    for i in range(len(words)):
        if not "'" in words[i]:
            if not "embed" in words[i]:
                if words[i] not in ["might", "also", "like", "lyric"]:
                    print(f"{words[i]:20} {counts[i]}")
                    counter += 1
        if counter >= 10:
            break

# get_most_common(1985)

####################
# PROBLEMS:
# "You might also like" (sometimes connected to something else, e.g. "lyricsyou")
# embed (sometimes connected to something else, e.g. "likeembed")
# n't, 's, 're, 'bout, c'mon etc. (GENERALLY: ')
# lyric or lyrics usually in the header of the files, not in the songtext
# ..., .., "", 
# Still a lot of meaningless words like "oooh", "ooooh", "ha", "huh"
# mr. instead of mister
# 
# 
# 
# 
# 
# 
####################
```

```python
for year in range(1955, 2024, 1):
    print(year)
    get_most_common(year)
    print("")
```
